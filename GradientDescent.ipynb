```
# Code is taken from CS 221 lecture vid 1
# All credits go to the professor
# This version contains my excessive but hopefully useful comments

#Let's say we have two points
points = [(2, 4), (4, 2)]

#We will define our cost function, J(w), which is F(w)= (xw-y)^2. 
# Have to do this for all x,y in points. And we want to sum it all up.

def J(w):
    return sum((w * x - y)**2 for x, y in points)

# To find gradient, we need to find the derivative of the cost function with respect to x:
def dJ(w):
    return sum(2*(w * x - y) * x for x, y in points)


# Gradient descent alogirthm

# Initalize model weights at random
w = 0
# eta is the step size (learning rate)

eta = 0.01

# we'll update the weights in the opposite of the gradient direction using equation w= w-eta*gradient
# reason we use a minus sign in this update equations is bc
# we want to move in the direction that decreases our cost function, which is the DRXN thatâ€™s opposite to the gradient.
# repeat this over many iterations using a for loop
for t in range(100):
    value = J(w)
    gradient = dJ(w)
    w = w - eta * gradient
    # print out these values, format them. on iteartion t, print out F(w) 
    print('iteration {}: w = {}, J(w) = {}, J\'(w)={}'.format(t, w, value, gradient))

    # code results: can see that w goes from 0.3 to about 0.78 and function value is going down(bc we're minimizing). Gradient 
    # is going to 0
    # Converges on the optimal solution is w=0.8. Slope is 0.8.
```
